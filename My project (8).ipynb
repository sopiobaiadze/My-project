```python
merged_tables.columns = [col.strip() for col in merged_tables.columns]
```

```python
merged_tables.columns=['date', 'total transaction Volume mlns', 'card payments_Volume', 'share %_Volume',
       'transactions value', 'card payments_Value', 'share %_Value', 'value per mln EU inhabitants']
```

```python
merged_tables['Non-card Payment Volume']=merged_tables['total transaction Volume mlns']-merged_tables['card payments_Volume']
```

```python
merged_tables.head()
```

```python
merged_tables['Non-card Payment Value']=merged_tables['transactions value']-merged_tables['card payments_Volume']
```

```python
merged_tables['Average card Transaction value']=merged_tables['card payments_Value']*1000/merged_tables['card payments_Volume']
```

```python
merged_tables['Average non-card Transaction value']=merged_tables['Non-card Payment Value']*1000/merged_tables['Non-card Payment Volume']
```

```python
merged_tables.head()
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
ax1 = sns.set_style(style=None, rc=None )
fig, ax1 = plt.subplots(figsize=(12,6))
sns.lineplot(data = merged_tables['Average card Transaction value'], marker='o', sort = False, ax=ax1, markers=True)
ax2 = ax1.twinx()
ax.set_title('card transactions')
sns.barplot(data = merged_tables, x='date', y='card payments_Volume', alpha=0.5, ax=ax2)
```

```python
fig, ax=plt.subplots()
ax.bar(merged_tables['date'], merged_tables['card payments_Volume'], label='card payments_Volume')
ax.bar(merged_tables['date'], merged_tables['Non-card Payment Volume'], bottom=merged_tables['card payments_Volume'],label='Non-card Payment Volume', color='green')
ax.set_title('payment transactions')
ax.legend()
```
